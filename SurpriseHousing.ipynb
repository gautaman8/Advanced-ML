### Results of Polynomial regression
R2 value is highest for Polynomial regression of degree 3 against 2,4,5.

We've to choose degree of 3. 


## Ridge Regression

```python
# Applying Ridge Regression with varying the hyperparameter 'lambda'

X_seq = np.linspace(top_X_train.min(),top_X_train.max(),300).reshape(-1,1) # values to be considered for predictor variable
lambdas = [0, 0.001, 0.01, 0.1, 1, 10, 100, 1000] # Higher the value of lambda, 
                                                  # more the regularization
for i in lambdas: # for each lambda we get different model coefficients
    degree = 3 # Degree for polynomial regression - chose 3 since this is the lowest number that gave a perfect fit
    # Creating degree 3 features
    ridgecoef = PolynomialFeatures(degree)
    # Transforming input features to polynomial features (1, x1, x2, x3, x4, x5)    
    X_poly = ridgecoef.fit_transform(top_X_train)
    ridgereg = Ridge(alpha = i) # Initialize the Ridge Regression model with a specific lambda
    ridgereg.fit(X_poly, y_train) # fit the model on the polynomial features
    #
    #Computing the r2 score
    y_pred = ridgereg.predict(ridgecoef.fit_transform(top_X_train))
    print("r2 score = " + str(r2_score(y_pred, y_pred))) 
    print(ridgereg.coef_) # model coefficients
```

## Lasso Regression

```python
# Applying Lasso Regression with varying the hyperparameter 'lambda'

lambdas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
for i in lambdas:
    degree = 3
    lassocoef = PolynomialFeatures(degree)
    # Transforming input features to polynomial features (1, x1, x2, x3, x4, x5)
    X_poly = lassocoef.fit_transform(top_X_train)
    lassoreg = Lasso(alpha = i)
    lassoreg.fit(X_poly, y_train)
    # Compute R^2 
    y_pred = lassoreg.predict(lassocoef.fit_transform(top_X_train))
    print("r2 score = " + str(r2_score(y_train, y_pred)))
    print(lassoreg.coef_)
```

### Ridge and Lasso Regression for specific lambda value

```python
# Get the model coefficients for specific lambda say 0.001

# Ridge Regression

ridgecoef = PolynomialFeatures(degree, include_bias = True) # Creating degree 3 features
# Transforming input features to polynomial features (1, x1, x2, x3, x4, x5)    
X_poly = ridgecoef.fit_transform(top_X_train)
ridgereg = Ridge(alpha = 0.001) # Initialize the Ridge Regression model with a specific lambda
ridgereg.fit(X_poly, y_train) # fit the model on the polynomial features
print(ridgereg.coef_)    
y_pred = ridgereg.predict(ridgecoef.fit_transform(top_X_train))
print("r2 score = " + str(r2_score(y_train, y_pred)))

# Lasso Regression
lassocoef = PolynomialFeatures(degree) # Creating degree 3 features
# Transforming input features to polynomial features (1, x1, x2, x3, x4, x5)
X_poly = lassocoef.fit_transform(top_X_train)
lassoreg = Lasso(alpha = 0.001)
lassoreg.fit(X_poly, y_train)
print(lassoreg.coef_)
y_pred = lassoreg.predict(lassocoef.fit_transform(top_X_train))
print("r2 score = " + str(r2_score(y_train, y_pred)))
```
