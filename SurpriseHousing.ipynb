# Data Cleaning
### Convert some integer features as categorical values 

```python
# ['id', 'MSSubClass'] should be string
data['Id'] = data['Id'].astype('string')
data['MSSubClass'] = data['MSSubClass'].astype('string')
```

### Clean missing values

```python
replace_na = {
    "LotFrontage" : 0, 
    "Alley" : "NA", 
    "MasVnrType" : "None", 
    "MasVnrArea" : 0, 
    "BsmtQual" : "NA", 
    "BsmtCond" : "NA", 
    "BsmtExposure" : "NA", 
    "BsmtFinType1" : "NA", 
    "BsmtFinType2" : "NA", 
    "FireplaceQu" : "NA", 
    "GarageType" : "NA", 
    "GarageYrBlt" : 0, 
    "GarageFinish" : "NA", 
    "GarageQual" : "NA", 
    "PoolQC" : "NA", 
    "Fence" : "NA", 
    "MiscFeature" : "NA"
}
for i in replace_na.keys(): 
    data[i].fillna(replace_na[i], inplace=True)

data.info()
# All the missing values are replaced appropriately.
```

### Dummy Variables

If categorical variables have more than two levels, they need to be converted to dummy variables. 


```python
# Create all dummy variables required and verify
var_list = ['MSSubClass','MSZoning', 'LandContour', 'LotConfig', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation', 'Heating', 'GarageType', 'MiscFeature','SaleType','SaleCondition']
for i in var_list: 
    exec(f"{i} = pd.get_dummies(data['{i}'], drop_first = True, prefix='{i}')")
    exec(f"print({i}.head())")



    
```

```python
# Add the results to the original dataframe
for i in var_list: 
    exec(f"data = pd.concat([data,{i}], axis = 1)")
```

## Step 4: Splitting the Data into Training and Testing Sets

As you know, the first basic step for regression is performing a train-test split.

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features

## Linear Regression model using RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
print(y_train.info())
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm,n_features_to_select=10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
print(col)
# Below are the predicted features for Linear regression using RFE
```
